```sql
create schema if not exists de_learning_workspace.audit;
```

```sql
CREATE TABLE IF NOT EXISTS de_learning_workspace.audit.pipeline_runs
(
    run_id STRING,
    pipeline_name STRING,
    start_time TIMESTAMP,
    end_time TIMESTAMP,
    status STRING,                  -- RUNNING, SUCCESS, FAILED
    trigger_type STRING,            -- MANUAL, SCHEDULED
    created_timestamp TIMESTAMP
)
USING DELTA;
```

```sql
CREATE TABLE IF NOT EXISTS de_learning_workspace.audit.pipeline_tasks
(
    run_id STRING,
    task_name STRING,
    layer STRING,                   -- Bronze, Silver, Gold
    start_time TIMESTAMP,
    end_time TIMESTAMP,
    status STRING,                  -- SUCCESS, FAILED
    rows_read BIGINT,
    rows_written BIGINT,
    error_message STRING,
    created_timestamp TIMESTAMP
)
USING DELTA;
```

```sql
CREATE TABLE IF NOT EXISTS de_learning_workspace.audit.watermark
(
    table_name STRING NOT NULL,
    last_processed_updated_at TIMESTAMP NOT NULL
)
USING DELTA;
```
