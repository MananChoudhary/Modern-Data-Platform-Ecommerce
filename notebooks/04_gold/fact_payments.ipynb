```sql
CREATE  TABLE if not EXISTS de_learning_workspace.gold.fact_payments
(
    payment_id BIGINT,
    order_id BIGINT,

    customer_sk BIGINT,

    payment_date DATE,
    payment_method STRING,
    payment_status STRING,

    amount DECIMAL(10,2),

    created_timestamp TIMESTAMP,
    updated_timestamp TIMESTAMP
)
USING DELTA;
```

```sql
INSERT INTO de_learning_workspace.gold.fact_payments
(
    payment_id,
    order_id,
    customer_sk,
    payment_date,
    payment_method,
    payment_status,
    amount,
    created_timestamp,
    updated_timestamp
)

SELECT

    p.payment_id,
    p.order_id,

    c.customer_sk,

    p.payment_date,
    p.payment_method,
    p.payment_status,

    p.amount,

    current_timestamp(),
    current_timestamp()

FROM de_learning_workspace.silver.payments p


JOIN de_learning_workspace.silver.orders o
ON p.order_id = o.order_id


JOIN de_learning_workspace.gold.dim_customer c
ON o.customer_id = c.customer_id
AND c.is_current = true


WHERE NOT EXISTS
(
    SELECT 1
    FROM de_learning_workspace.gold.fact_payments f
    WHERE f.payment_id = p.payment_id
);
```
