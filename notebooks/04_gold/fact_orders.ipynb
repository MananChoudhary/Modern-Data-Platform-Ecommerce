```python
%sql

CREATE  table if not EXISTS de_learning_workspace.gold.fact_orders
(
    order_id INT,

    customer_sk BIGINT,

    order_date TIMESTAMP,
    order_status STRING,
    payment_status STRING,

    total_amount DECIMAL(12,2),

    created_timestamp TIMESTAMP,
    updated_timestamp TIMESTAMP
)
USING DELTA;
```

```python
%sql

INSERT INTO de_learning_workspace.gold.fact_orders
(
    order_id,
    customer_sk,
    order_date,
    order_status,
    payment_status,
    total_amount,
    created_timestamp,
    updated_timestamp
)

SELECT

    o.order_id,

    c.customer_sk,

    o.order_date,
    o.order_status,
    o.payment_status,

    o.total_amount,

    current_timestamp(),
    current_timestamp()

FROM de_learning_workspace.silver.orders o

INNER JOIN de_learning_workspace.gold.dim_customer c
ON o.customer_id = c.customer_id
AND c.is_current = TRUE

WHERE NOT EXISTS
(
    SELECT 1
    FROM de_learning_workspace.gold.fact_orders f
    WHERE f.order_id = o.order_id
);
```
