```python
%sql
CREATE or replace TABLE gold.fact_inventory_snapshot
(
    inventory_snapshot_date DATE,

    inventory_id BIGINT,

    product_sk BIGINT,

    warehouse_location STRING,

    available_quantity INT,

    reserved_quantity INT,

    inventory_status STRING,

    source_updated_at TIMESTAMP,

    created_timestamp TIMESTAMP
)
USING DELTA;
```

```python
%sql
INSERT INTO de_learning_workspace.gold.fact_inventory_snapshot
(
    inventory_snapshot_date,
    inventory_id,
    product_sk,
    warehouse_location,
    available_quantity,
    reserved_quantity,
    inventory_status,
    source_updated_at,
    created_timestamp
)

SELECT

    current_date(),

    i.inventory_id,

    p.product_sk,

    i.warehouse_location,

    i.available_quantity,

    i.reserved_quantity,

    i.inventory_status,

    i.updated_at,

    current_timestamp()

FROM silver.inventory i


JOIN gold.dim_product p
ON i.product_id = p.product_id
AND p.is_current = true;
```
